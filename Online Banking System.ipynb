```python
#### grouping data by various time segments.convert our date column to datetime, which will split up the date objects into separate data points of year, month, and day. 
```

```python
##convert date column to date time 

df["date"]= pd.to_datetime(df["date"])
```

```python
### which month has the most lightening srike 


# step 1: create a new column that we call "month"

###### dt.month extracts just the month information (as a numeric value) from the date. 

import datetime as dt

df["month"]=df["date"].dt.month
df.head()
```

```python
df.tail()
```

```python

#### adding one column that we call "year " to the dataset
#### dt.year extracts just the year information (as a numeric value) from the date. 

import datetime as dt

df["year"]=df["date"].dt.year
df.head()
```

```python
#### Calculate the number of strikes per month

###### Now, we will sort our values by most strikes per month. Use groupby(), sum() and sort_values() from pandas again.



df.groupby(['month']).sum().sort_values('number_of_strikes', ascending=False).head(12)
```

```python

df.sort_values('number_of_strikes', ascending=False).groupby(['month']).sum().head(12)
```

```python
#### dt.day extracts just the month information (as a numeric value) from the date. 
df["day"]=df["date"].dt.day
df.head()
```

```python
df.tail()
```

```python
# using function dt.month.name() in order to adding a new column named 'month_txt' to the DataFrame df,
#this new column will contain the month names corresponding to the values in the 'date' column.

df['month_txt'] = df['date'].dt.month_name()
df.head()
```
